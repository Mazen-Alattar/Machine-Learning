## Bar chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,85,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
# we should plot the relation between numneric data , company is string so, we mapped company values into numpy array 
#after plotting we presented values of company not from the numpy arrays 

plt.bar(xpos,revenue,label='Revenue')
plt.xticks(xpos,company)      # mapping xpos to comapny values ( 0 --> Googl , 1 -- > Amzn)
plt.title('US Technology Stocks')
plt.xlabel('Company')
plt.ylabel('Revenue(Bln)')
plt.legend()
plt.show()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.4,profit,width=0.4,label='Profit')
plt.xticks(xpos,company)
plt.title('US Technology Stocks')
plt.xlabel('Company')
plt.ylabel('Revenue(Bln)')
plt.legend()
plt.show()
```

## Horizontal bar chart using barh fun

```python
plt.barh(xpos,revenue,label='Revenue')
plt.yticks(xpos,company)      
plt.title('US Technology Stocks')
plt.ylabel('Company')
plt.xlabel('Revenue(Bln)')
plt.legend()
plt.show()
```

## Histogram

**We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic**

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

plt.hist(blood_sugar,rwidth=0.8,color='green')     # by default number of bins is set to 10
plt.show()
```

```python
plt.hist(blood_sugar,rwidth=0.5,bins=3)
plt.show()
```

<h3 style='color:black'>Mutiple data samples in a histogram</h3>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women],color=['b','deeppink'],bins=[77,100,125,150], rwidth=0.8,label=['Men','Women'])
plt.legend()
plt.show()
```

<h3 style='color:black'>histtype=step</h3>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
plt.show()
```

<h3 style='color:black'>horizontal orientation</h3>

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
plt.show()
```

## Pie Chart

```python
exp_vals = [1400,600,300,300,250] #expenses 
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

<h3 style="color:black">Show percentages for every pie. Specify radius to increase chart size and shadow</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,shadow=True,radius=1.66,autopct='%1.1f%%')
plt.show()
```

<h3 style="color:black">Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0,0.3])
plt.show()  
```

<h3 style="color:black">counterclock and angle properties</h3>

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,counterclock=True,autopct='%1.1f%%',explode=[0,0,0,.1,0.3],startangle=180)
plt.show()
```

<h2 style="color:green" align="center">Saving chart to a file using savefig</h2>

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,counterclock=True,autopct='%1.1f%%',explode=[0,0,0,.1,0.3],startangle=180)
plt.savefig("piechart.png", bbox_inches="tight", pad_inches=7, transparent=True)
plt.show()
```

<h2>You can also save in pdf format and to a different location on your computer by specifying absolute path<h2>

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("D:\piechart.pdf", bbox_inches="tight", pad_inches=1, transparent=True)
```
